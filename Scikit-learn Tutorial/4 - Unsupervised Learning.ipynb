```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

### Generate a Toy Dataset

```python
# Create a toy dataset
X, _ = make_blobs(n_samples=1700, centers=7, cluster_std=4.90, random_state=0)
print(X.shape)

# plot these blobs
plt.style.use('dark_background')
plt.scatter(X[:, 0], X[:, 1], s=10)
plt.axis('equal')
plt.show()
```

### Apply K-Means on it

Read over K-Means documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```python
# Create subplots for visualizing clusters at each iteration
fig, axs = plt.subplots(1, 4, figsize=(15, 3))
axs = axs.flatten()

for i in range(4):
    # Fit K-means and get cluster labels
    kmeans = KMeans(n_clusters=4, init='random', max_iter=i+1, random_state=0, n_init=10)
    kmeans.fit(X)                       # apply K-mean here!
    labels = kmeans.labels_             # Get the label (i.e., cluster) of each point

    # Get centroids and store for visualization
    centroids = kmeans.cluster_centers_ # Get the centroids
    
    # Plot the data points and centroids
    # increase dpi
    fig.set_dpi(300)
    axs[i].scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.9)
    #axs[i].scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, marker='X')
    axs[i].set_title(f"Iteration {i+1}")

plt.tight_layout()
plt.show()
```

More or less converged from the third iteration
